```python
#Q1.
print('''We use for loop functions when we have to iterate each value of the given condition 
For example: We'll be creating the multiple of 9:''')

ls = [1, 2, 3, 4, 5, 6, 7, 9, 10] # We are creating a list & assigning it as ls 

for i in ls: # This line of code explains that we are iterating each element in list(ls) by using variable i 
    print(i * 9) #Now it'll print the multiples of 9
    
print('\n')

print('''While loop with Example
We use while loop functions, when we have to iterate the values until the specified condition become True
For example: We'll be generating numbers from 1 to 100''')

p = 1 #We are creating a variable 'p' whose value is 1
while  p <= 50: #Now, We are creating a while loop whose condition are if p is less than or equal to 50, print all the value of p with adding 1 unless the value becomes True
    print(p)
    p += 1
```

```python
#Q2.Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

#Creating the sum and product of first 10 natural numbers using for loops.
sums = 0 #Here, We are assigning variable(sum) as 0
prod = 1 #Assigning variable(prod) as 1

for i in range(1,10+1): # Now, we're creating a for loop function of variable (i) by giving a range of 1 to 10, Note: We wrote 10+1 as per the upperbound + 1 rule. 
    sums +=  i #We are over-assigning the variable for every number in the given range we will add the previous value of sum with the iterated value
    prod *=  i #We are over-assigning the variable for every number in the given range we will multiply the previous value of sum with the iterated value
print(f"The sum of first 10 natural numbers will be {sums}.") 
print(f"The Product of first 10 natural numbers will be {prod}.")
# Creating the sum of first 10 natural numbers using a while loop
# Sum
sumW = 0 
crN = 1  # Initialize the current number to 1
while crN <= 10:  # Continue until the current number reaches 10
    sumW += crN  # Add the current number to the sum
    crN += 1  # Move to the next number

# Print the sum
print("Sum of first 10 natural numbers using while loop:", sumW)

# Creating the product of first 10 natural numbers using a while loop
# Product
pdablu = 1  # Initialize the product to 1 (identity element for multiplication)
prod = 1  # Initialize the current number to 1

while prod <= 10:  # Continue until the current number reaches 10
    pdablu *= prod  # Multiply the product by the current number
    prod += 1  # Move to the next number

# Print the product
print("Product of first 10 natural numbers using while loop:", pdablu)
```

```python
# Q3. Create a python program to compute the electricity bill for a household.
# The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
# unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
# be charged Rs. 10 per unit, After 300 units and above, the user will be charged Rs. 20 per unit.
# You are required to take the units of electricity consumed in a month from the user as input.
# Your program must pass this test case: when the unit of electricity consumed by the user in a month is 310, the total electricity bill should be 2250.

# Taking units of electricity consumed as input from the user
unit = int(input())

# Initializing charges for each block to zero
charge1 = charge2 = charge3 = charge4 = 0

# Checking the range of units consumed to calculate charges for each block
if unit == 100:
    charge1 = 4.5 * unit
elif unit > 100 and unit <= 200:
    charge1 = 4.5 * 100
    charge2 = 6 * (unit - 100)
elif unit > 200 and unit <= 300:
    charge1 = 4.5 * 100
    charge2 = 6 * 100
    charge3 = 10 * (unit - 200)
elif unit > 300:
    charge1 = 4.5 * 100
    charge2 = 6 * 100
    charge3 = 10 * 100
    charge4 = 20 * (unit - 300)

# Calculating the total electricity charges
total_charges = charge1 + charge2 + charge3 + charge4

# Displaying the total charges as an integer
print(int(total_charges))
```

```python
# #Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print that list.

# #Using for loops
cuber_1 = []
for i in range(1, 100+1):
    c = i ** 3
    if c%4 == 0 or c%5 == 0 :
        cuber_1.append(i)
print(cuber_1)

#Using while loop:
cuber_2 = []
# ct = 1
count = 1
while count <= 100:
    ct = count ** 3 
    count += 1
    if ct%4 == 0 or ct%5 == 0:
        cuber_2.append(count-1)
print(cuber_2)
```
